<a class="heading-toc" id="types-of-data-analytics"></a> <h2 id="mntl-sc-block_1-0-26" class="comp mntl-sc-block finance-sc-block-heading mntl-sc-block-heading"> <span class="mntl-sc-block-heading__text"> Types of Data Analytics </span> </h2>
<p id="mntl-sc-block_1-0-27" class="comp mntl-sc-block finance-sc-block-html mntl-sc-block-html">
Data analytics is broken down into four basic types.
</p>
<div id="mntl-sc-block_1-0-28" class="comp mntl-sc-block mntl-sc-block-adslot mntl-block"></div>
<ol id="mntl-sc-block_1-0-29" class="comp mntl-sc-block finance-sc-block-html mntl-sc-block-html">
<li><strong>Descriptive analytics</strong> describes what has happened over a given period of time. Have the number of views gone up? Are sales stronger this month than last?</li><li><strong>Diagnostic analytics</strong> focuses more on why something happened. This involves more diverse data inputs and a bit of hypothesizing. Did the weather affect beer sales? Did that latest marketing campaign impact sales?</li><li><strong>Predictive analytics</strong> moves to what is likely going to happen in the near term. What happened to sales the last time we had a hot summer? How many weather models predict a hot summer this year?</li><li><strong>Prescriptive analytics</strong> suggests a course of action. If the likelihood of a hot summer is measured as an average of these five weather models is above 58%, we should add an evening shift to the brewery and rent an additional tank to increase output.</li>
</ol>
<div id="mntl-sc-block_1-0-30" class="comp mntl-sc-block mntl-sc-block-adslot mntl-block"></div>
<p id="mntl-sc-block_1-0-31" class="comp mntl-sc-block finance-sc-block-html mntl-sc-block-html">
Data analytics underpins many quality control systems in the financial world, including the ever-popular <a href="https://www.investopedia.com/terms/s/six-sigma.asp" data-component="link" data-source="inlineLink" data-type="internalLink" data-ordinal="1">Six Sigma</a> program. If you aren’t properly measuring something—whether it's your weight or the number of defects per million in a production line—it is nearly impossible to optimize it.
</p>

## Role of a Data Analyst

A Data Analyst interprets data and turns it into information which can offer ways to improve a business, thus affecting business decisions. Data Analysts gather information from various sources and interpret patterns and trends – as such a Data Analyst job description should highlight the analytical nature of the role. Once data has been gathered and interpreted, the Data Analyst will report back what has been found in a comprehensive study to the wider business/relevant colleagues.

<p>Data Analysts often make recommendations about the methods and ways in which a company obtains and analyses data to improve quality and the efficiency of data systems. A Data Analyst job description should include, but not be limited to:</p>

<ul><li>Collecting and interpreting data</li>
<li>Analysing results</li>
<li>Reporting the results back to the relevant members of the business</li>
<li>Identifying patterns and trends in data sets</li>
<li><a href="https://www.roberthalf.com.au/career-advice/career-development/teamwork-skills" title="Working alongside teams">Working alongside teams</a> within the business or the management team to establish business needs</li>
<li>Defining new data collection and analysis processes</li>
</ul>

**If we notice there are duplicates still left over,that is the `INR` is repeating twice,if you run this query in the MYSQL workbench,i.e this one `select distinct(transactions.currency) from sales.transactions;`,you'' find 4 values,which must be 2,but are 4 i.e there're duplicates,so we've to remove them tooo,the problem is one `INR` is just INR,but the other `INR` is of this type: `INR\r` type,so we need to remove this kind,so when i ran this in SQL : `select count(*) from sales.transactions where transactions.currency='INR\r';`,the count was 15k!!,now when you run this : `select count(*) from transactions where currency = 'INR';`,the count comes out to be 279,that means that records with `INR\r` are good records as they're in majority and the ones in `INR` are bad as they're in minority,this is what i'm thinking,this may differ from client to client,now in case of `USD` also we'vve USD\r and USD,but here USD\r is defective and USD is normal,so i'm removing all the records which are having `INR` and `USD` and keeping only `INR\r and USD\r` as they're in majority**

**so in the PowerBI super query write this : `= Table.SelectRows(#"Filtered Rows", each ([currency] = "INR#(cr)" or [currency] = "USD#(cr)"))`**
